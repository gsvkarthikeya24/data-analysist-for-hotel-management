## Data loading

### Subtask:
Load the dataset "Dataset  hotels types and level  of mangement .csv" into a pandas DataFrame.


**Reasoning**:
Load the dataset into a pandas DataFrame and display the first few rows to verify.



```python
import pandas as pd

try:
    df = pd.read_csv('Dataset  hotels types and level  of mangement .csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'Dataset  hotels types and level  of mangement .csv' not found.")
    df = None
except pd.errors.ParserError:
    try:
        df = pd.read_csv('Dataset  hotels types and level  of mangement .csv', encoding='latin-1')
        display(df.head())
    except:
        print("Error: Unable to parse the CSV file with default or latin-1 encoding.")
        df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the loaded DataFrame (`df`) to understand its structure, identify key variables, and determine the shape and distribution of the data.


**Reasoning**:
I need to explore the DataFrame's structure, inspect data types, identify key variables, check for missing values, and analyze the distribution of key variables.



```python
# Examine the DataFrame's structure
print("Shape of the DataFrame:", df.shape)
display(df.head())
display(df.tail())
print("Column names:", df.columns.tolist())

# Inspect data types
print("\nData types of each column:")
print(df.dtypes)

# Identify key variables (based on the provided column names and context)
# Assuming 'Cuisines' represents hotel types and 'Price range' represents management level.
key_variables = ['Cuisines', 'Price range']
print("\nKey variables:", key_variables)

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nPercentage of missing values per column:")
print((df.isnull().sum() / len(df)) * 100)

# Analyze the distribution of key variables
print("\nDescriptive statistics for 'Price range' (numerical):")
print(df['Price range'].describe())

print("\nValue counts for 'Cuisines' (categorical):")
print(df['Cuisines'].value_counts())
```

## Data cleaning

### Subtask:
Clean the DataFrame `df` by handling missing values in the 'Cuisines' column and removing any duplicate rows.


**Reasoning**:
Fill missing values in the 'Cuisines' column with "Unknown", remove duplicate rows, and reset the index. Then, verify the changes by printing the number of missing values, the shape of the DataFrame, and the first 5 rows.



```python
# Fill missing values in 'Cuisines' with "Unknown"
df['Cuisines'].fillna("Unknown", inplace=True)

# Remove duplicate rows and reset the index
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

# Verify the changes
print("Number of missing values in 'Cuisines' after filling:", df['Cuisines'].isnull().sum())
print("Shape of the DataFrame after removing duplicates:", df.shape)
display(df.head())
```

## Data wrangling

### Subtask:
Transform the data in DataFrame `df` to prepare it for analysis and visualization.  Specifically, create a new categorical feature based on the 'Price range' column, representing hotel price categories.


**Reasoning**:
Analyze the distribution of 'Price range' and create a new 'Price Category' column based on quartiles.



```python
import matplotlib.pyplot as plt

# Analyze the distribution of 'Price range'
price_range_description = df['Price range'].describe()
print(price_range_description)

plt.figure(figsize=(8, 6))
plt.hist(df['Price range'], bins=range(int(df['Price range'].min()), int(df['Price range'].max()) + 2), align='left', rwidth=0.8)
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.title('Distribution of Price Range')
plt.show()

# Categorize hotels based on quartiles
quartiles = df['Price range'].quantile([0.25, 0.5, 0.75])
print("Quartiles:", quartiles)

def categorize_price(price):
    if price <= quartiles[0.25]:
        return 'Budget'
    elif price <= quartiles[0.5]:
        return 'Mid-range'
    elif price <= quartiles[0.75]:
        return 'Above Mid-range'
    else:
        return 'Luxury'

df['Price Category'] = df['Price range'].apply(categorize_price)

# Verify the transformation
display(df.head())
print(df['Price Category'].value_counts())
```
